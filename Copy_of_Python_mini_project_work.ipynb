```python
result['Daily Return'] = result['Adj Close'].pct_change()
result.head(10)
```

```python
def trend(x):
  if x > -0.015 and x <= 0.015:
    return 'Slight or No change'
  elif x > 0.015 and x <= 0.04:
    return 'Slight Positive'
  elif x < -0.015 and x >= -0.4:
    return 'Slight Negative'
  elif x > 0.04 and x <= 0.06:
    return 'Positive'
  elif x < -0.04 and x >= -0.06:
    return 'Negative'
  elif x > 0.06 and x <= 0.07:
    return 'Among top gainers'
  elif x < -0.06 and x >= -0.07:
    return 'Among top losers'
  elif x > 0.07:
    return 'Bull run'
  elif x <= -0.07:
    return 'Bear drop'
```

```python
result['Trend'] = result['Daily Return'].apply(lambda x: trend(x))
result.head(10)
```

```python
for company_name in result.Company.unique():
    df = result[result['Company']==company_name]
    values = df['Trend'].value_counts().values
    label = df['Trend'].value_counts().index
    patches, texts = plt.pie(values, labels = label, startangle = 35, rotatelabels = True, labeldistance = 1.05, radius = 5)
    plt.title(company_name, color = 'black', loc = 'center', fontweight = "bold")
    plt.show()
```
